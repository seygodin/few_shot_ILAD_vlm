```python
import pandas as pd
import re

def get_result(file_path, data_name):
    # 결과를 저장할 리스트
    best_la_auc_values = []

    # 파일을 읽고 각 라인을 처리
    with open(file_path, 'r') as file:
        for line in file:
            # 'Best LA AUC' 값을 찾는다
            match = re.search(r'Best LA AUC: ([0-9.]+)', line)
            if match:
                # 찾은 값을 리스트에 추가한다
                best_la_auc_values.append(float(match.group(1)))

    # 리스트를 사용하여 pandas DataFrame을 생성
    df = pd.DataFrame(best_la_auc_values, columns=[data_name])
    return df

import os

def gather_result(data_name):
    templates_path = ["_1_f_f_result", "_1_t_f_result", "_1_f_t_result", "_1_t_t_result",
                "_5_f_f_result", "_5_t_f_result", "_5_f_t_result", "_5_t_t_result"]
    templates_name = ["_1", "_1_mask", "_1_text", "_1_mask_text",
                      "_5", "_5_mask", "_5_text", "_5_mask_text"]
    df_list = []
    for template_name, template_path in zip(templates_name, templates_path):
        path = data_name + template_path
        name = data_name + template_name

        if not os.path.exists(path):
            continue
        df = get_result(path, name)

        df_list.append(df)

    final_df = pd.concat((df_list), axis=1)
    
    return final_df

def show_result(dir_name):
    data_names = ["breakfast", "banana_juice", "orange_juice", "cherry_juice", "pushpins",
                  "screw_bag", "blue_splicing", "red_splicing", "yellow_splicing"]
    
    result = {}
    
    for data_name in data_names:
        data_path = dir_name + "/" + data_name
        data_df = gather_result(data_path)
        avg_perfo = data_df.mean().tolist()
        result[data_name] = avg_perfo
    
    df = pd.DataFrame(result)
    #df.index = ["1", "1_mask", "1_text", "1_mask_text",
    #                  "5", "5_mask", "5_text", "5_mask_text"]
    
    df.index = ["1_mask","5_mask"]
    return df

def add_avg_perfo(df):
    avg_perfo = df.mean().tolist()
    df.loc[len(df.index)] = avg_perfo
    df.index = df.index.tolist()[:len(df.index)-1] + ["Average"]
    return df

def compare_two_result(df1, df1_name, df2, df2_name):
    #df1 and df2 must share column names and index names
    result = {}
    for col in df1.columns:
        result[col + " " + df1_name] = df1[col]
        result[col + " " + df2_name] = df2[col]
        result["Delta " + col] = df2[col] - df1[col] 
    df = pd.DataFrame(result)
    return df
```
