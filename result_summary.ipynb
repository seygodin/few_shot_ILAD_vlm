```python
import numpy as np
from scipy import stats

def get_p_value(accuracies):
    mean_accuracy = np.mean(accuracies)
    std_accuracy = np.std(accuracies, ddof=1)
    n = len(accuracies)

    # 표준오차 계산
    standard_error = std_accuracy / np.sqrt(n)
    # 기준 정확도 (예: 랜덤 추측의 정확도)
    baseline_accuracy = 0.5
    # t-검정 수행
    t_stat, p_value = stats.ttest_1samp(accuracies, baseline_accuracy)
    return p_value
```

```python
def get_result(file_path, data_name):
    # 결과를 저장할 리스트
    best_la_auc_values = []

    # 파일을 읽고 각 라인을 처리
    with open(file_path, 'r') as file:
        for line in file:
            # 'Best LA AUC' 값을 찾는다
            match = re.search(r'Best LA AUC: ([0-9.]+)', line)
            if match:
                # 찾은 값을 리스트에 추가한다
                best_la_auc_values.append(float(match.group(1)))

    # 리스트를 사용하여 pandas DataFrame을 생성
    df = pd.DataFrame(best_la_auc_values, columns=[data_name])
    return df
```

```python
import os

def gather_result(data_name):
    templates_path = ["_1_f_f_result", "_1_t_f_result", "_1_f_t_result", "_1_t_t_result",
                "_5_f_f_result", "_5_t_f_result", "_5_f_t_result", "_5_t_t_result"]
    templates_name = ["_1", "_1_mask", "_1_text", "_1_mask_text",
                      "_5", "_5_mask", "_5_text", "_5_mask_text"]
    df_list = []
    for template_name, template_path in zip(templates_name, templates_path):
        path = data_name + template_path
        name = data_name + template_name

        if not os.path.exists(path):
            continue
        df = get_result(path, name)

        df_list.append(df)

    final_df = pd.concat((df_list), axis=1)
    
    return final_df
```

```python
results = {}

data_names = ["breakfast", "screw_bag", "pushpins", "banana_juice", "orange_juice", "cherry_juice", "red_splicing", "blue_splicing", "yellow_splicing"]

for data_name in data_names:
    df = gather_result(f"paper_results/{data_name}")
    one_col_namme = f"paper_results/{data_name}_1_mask_text"
    five_col_namme = f"paper_results/{data_name}_5_mask_text"

    one_shot = df[one_col_namme]
    five_shot = df[five_col_namme]
    p_value_1 = get_p_value(one_shot)
    p_value_5 = get_p_value(five_shot)
    results[data_name] = [p_value_1, p_value_5]
result_df = pd.DataFrame(results).T
result_df.columns = ["1-shot", "5-shot"]
result_df
```

```python
import numpy as np
from sklearn.metrics import precision_recall_curve, auc

def calculate_auprc_and_optimal_f1(true_label, pred_score):
    # Precision-Recall 곡선 계산
    precision, recall, thresholds = precision_recall_curve(true_label, pred_score)
    # AUPRC 계산
    auprc = auc(recall, precision)
    
    # F1-스코어 계산 (각 임계값에서)
    f1_scores = 2 * (precision[:-1] * recall[:-1]) / (precision[:-1] + recall[:-1])
    # NaN 값을 0으로 대체
    f1_scores = np.nan_to_num(f1_scores)
    # 최적의 F1-스코어 인덱스 찾기
    optimal_idx = np.argmax(f1_scores)
    # 최적 임계값
    optimal_threshold = thresholds[optimal_idx]
    # 최적 F1-스코어
    optimal_f1 = f1_scores[optimal_idx]
    
    print("Precision-Recall 곡선 아래 면적 (AUPRC):", auprc)
    print("최적 임계값:", optimal_threshold)
    print("최적 임계값에서의 F1-스코어:", optimal_f1)
    
    return auprc, optimal_f1, optimal_threshold

```
